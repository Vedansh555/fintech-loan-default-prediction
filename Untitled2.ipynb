```python
import pandas as pd
df = pd.read_csv("medium_churn_1000.csv")
print(df.head(2))
```

```python
print("shape",df.shape)
print(df.info())
print(df.isnull().sum())
print(df["Churn"].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["MonthlyCharges"],kde = True)
plt.title("Monthly charges distribution")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder , MinMaxScaler
scaler = MinMaxScaler()

# Encode categorical features
le = LabelEncoder()
df = pd.get_dummies(df, columns=['Gender', 'Contract'], drop_first=True)
df['Churn'] = df['Churn'].map({'No': 0, 'Yes': 1})  # Convert to binary

numeric_cols = ["Age","Tenure","MonthlyCharges","TotalCharges"]
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
# Features and label
X = df.drop(['CustomerID', 'Churn'], axis=1)
y = df['Churn']

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
from imblearn.over_sampling import SMOTE

# Balance the classes
sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
lr = LogisticRegression(max_iter=200)
lr.fit(X_train,y_train)

rf = RandomForestClassifier(n_estimators=100,random_state = 42)
rf.fit(X_train,y_train)

from xgboost import XGBClassifier
model = XGBClassifier(eval_metric='logloss')
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.ensemble import VotingClassifier
# Logistic Regression
lr_preds = lr.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, lr_preds))
print(classification_report(y_test, lr_preds))

# Random Forest
rf_preds = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, rf_preds))
print(classification_report(y_test, rf_preds))

# XGB

model_preds = model.predict(X_test)
print("XGB Accuracy:", accuracy_score(y_test, model_preds))
print(classification_report(y_test, model_preds))

ensemble = VotingClassifier(
    estimators=[ ('lr', lr), ('rf', rf),('xgb', model)],
    voting='soft'  # Use probabilities instead of just votes
)
ensemble.fit(X_train, y_train)
ensemble_preds = ensemble.predict(X_test)
print("stack Accuracy:", accuracy_score(y_test, ensemble_preds))

print(classification_report(y_test, model_preds))
# Optional: confusion matrix
sns.heatmap(confusion_matrix(y_test, rf_preds), annot=True, fmt='d', cmap='Blues')
plt.title("Random Forest Confusion Matrix")
plt.show()

```

```python
results = X_test.copy()
results['Actual'] = y_test
results['Predicted_RF'] = rf_preds
results['Predicted_LR'] = lr_preds
results["Pridicted XGB"] = model_preds
results["Predicted stack"] = ensemble_preds
results.to_csv("fintech_predictions.csv", index=False)

print("Saved predictions to fintech_predictions.csv")
```
